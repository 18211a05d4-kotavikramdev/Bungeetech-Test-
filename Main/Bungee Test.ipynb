```python
import pandas as pd
data = pd.read_csv('../input/question-1/main.csv')
del data["Total"]
population = data["Population"]
popul = []
for i in range(9, len(population) + 1, 10):
    popul.append(population.iloc[i])
popul.append(int(population.iloc[-1]))
del data["Population"]
data = data.groupby((data.Year//10)*10).sum()
data.insert(loc = 1, column = 'Population', value = popul)
del data["Year"]
print(data)
```

```python
import pandas as pd
data = pd.read_csv('../input/question-2/main.csv')
data = data[["age", "occupation"]]
print(data)
```

```python
result = data.groupby('occupation').agg({'age': ['min', 'max']})
print(result)
```

```python
result.to_csv('main.csv')
```

```python
import pandas as pd
data = pd.read_csv('../input/question-3/main.csv')
data = data[["Team", "Yellow Cards", "Red Cards"]]
print(data)
```

```python
result = data.sort_values(by = ['Red Cards', 'Yellow Cards'], ascending = False)
print(result)
```
